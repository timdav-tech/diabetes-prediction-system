```python
import joblib
import numpy as np
import pandas as pd
# ── Load saved model and scaler ──
model = joblib.load('diabetes_model.pkl')
scaler = joblib.load('diabetes_scaler.pkl')
print('=== DIABETES PREDICTION SYSTEM ===')
print('Enter patient data below (press Enter after each value):')
# ── Collect input from user ──
try:
 pregnancies = float(input('Number of pregnancies (e.g. 2): '))
 glucose = float(input('Glucose level mg/dL (e.g. 120): '))
 blood_press = float(input('Blood pressure mm Hg (e.g. 70): '))
 skin_thick = float(input('Skin thickness mm (e.g. 20): '))
 insulin = float(input('Insulin mu U/ml (e.g. 80): '))
 bmi = float(input('BMI kg/m² (e.g. 25.0): '))
 dpf = float(input('Diabetes pedigree function (e.g. 0.5):'))
 age = float(input('Age in years (e.g. 35): '))
except ValueError:
 print('ERROR: Please enter numeric values only.')
 exit()
    
# ── Prepare input ──
patient = pd.DataFrame({
    'Pregnancies': [pregnancies],
    'Glucose': [glucose],
    'BloodPressure': [blood_press],
    'SkinThickness': [skin_thick],
    'Insulin': [insulin],
    'BMI': [bmi],
    'DiabetesPedigreeFunction': [dpf],
    'Age': [age]
})
# ── Scale using the SAVED scaler ──
patient_scaled = scaler.transform(patient)
# ── Predict ──
prediction = model.predict(patient_scaled)[0]
probability = model.predict_proba(patient_scaled)[0]
# ── Display Result ──
print('\n' + '='*45)
print(' PREDICTION RESULT')
print('='*45)
if prediction == 1:
 print(' RESULT : ⚠️ DIABETIC (Positive)')
 print(f' CONFIDENCE: {probability[1]*100:.1f}% likely diabetic')
 print('\n RECOMMENDATION: Please consult a doctor')
 print(' immediately for proper diagnosis.')
else:
 print(' RESULT : ✅ NOT DIABETIC (Negative)')
 print(f' CONFIDENCE: {probability[0]*100:.1f}% likely healthy')
 print('\n RECOMMENDATION: Maintain healthy lifestyle.')
 print(' Regular checkups recommended.')
print('='*45)
print('\nProbabilities:')
print(f' Not Diabetic: {probability[0]*100:.1f}%')
print(f' Diabetic: {probability[1]*100:.1f}%')
print('\nNOTE: This is an ML prediction, NOT medical advice.')
print('Always consult a qualified healthcare professional.')
```
